```python
# subsampling 4 = taking 50% data
SAMPLING_RATE = 1/2

# Downsample by users
user_id_unique = interactions_train_active.user_id.unique()
user_id_sample = pd.DataFrame(user_id_unique, columns=['unique_user_id']).sample(frac= SAMPLING_RATE, replace=False, random_state=1)

ratings_sample_50 = interactions_train_active.merge(user_id_sample, left_on='user_id', right_on='unique_user_id').drop(['unique_user_id'], axis=1)
print(ratings_sample_50.head())
print(ratings_sample_50.shape)
```

```python
ratings_sample_50['rating'].value_counts()
```

**Replacing ratings integer as the following.**

```python
ratings_sample_50['rating'] = ratings_sample_50['rating'].replace(5,6)
ratings_sample_50['rating'] = ratings_sample_50['rating'].replace(4,5)
ratings_sample_50['rating'] = ratings_sample_50['rating'].replace(3,4)
ratings_sample_50['rating'] = ratings_sample_50['rating'].replace(2,3)
ratings_sample_50['rating'] = ratings_sample_50['rating'].replace(1,2)
ratings_sample_50['rating'] = ratings_sample_50['rating'].replace(0,1)
```

```python
user_item_m_50 = ratings_sample_50.pivot('user_id','recipe_id','rating').fillna(0)
print(f'Shape: {user_item_m_50.shape}')
```

```python
user_item_m_50.head(5)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# cosine similarity between users
X_user = cosine_similarity(user_item_m_50) # user similarity matrix
```

```python
X_user
```

```python
def recipe_recommender(user_item_m_50, X_user, user, k=20, top_n=10):
    user_ix = user_item_m_50.index.get_loc(user)
    print(user_ix)
    
    user_similarities = X_user[user_ix]
    print(user_similarities)
    
    most_similar_users = user_item_m_50.index[user_similarities.argpartition(-k)[-k:]]
    print(f"the most similar users \n {most_similar_users}")
    
    recommend_food = user_item_m_50.loc[most_similar_users].mean(0).sort_values(ascending=False)
    
    m_rated_recipes = user_item_m_50.loc[user].gt(0)
    print(f"rated recipes \n {m_rated_recipes}")
    
    rated_recipes = m_rated_recipes.index[m_rated_recipes].tolist()
    print(f"recorded recipes \n {rated_recipes}")
    
    recommend_food_top = recommend_food.drop(rated_recipes).head(top_n)
    return recommend_food_top.index.to_frame().reset_index(drop=True).merge(raw_recipes)
```

```python
recommendation = recipe_recommender(user_item_m_50, X_user, user = 1634)
```

```python
user_item_m_05
```

```python
# cosine similarity between items
from sklearn.metrics.pairwise import cosine_similarity
X_items = cosine_similarity(user_item_m_05.T) # item similarity matrix
```

```python
X_items
```

```python
X_items.shape
```

```python
def item_based(item, similar, top_rank):
    # find the recipe_id and the corresponding coloumn
    locate = user_item_m_05.columns.get_loc(item)
    sim = X_items[locate]
    # get the most similar neigbours
    most_similar = user_item_m_05.columns[sim.argpartition(-(similar+1))[-(similar+1):]]
    # return the top rank list.
    return (most_similar.difference([item]).to_frame().reset_index(drop=True).head(top_rank))
```
